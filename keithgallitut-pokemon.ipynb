## Filtering Data

```python
# only display type 1 grass type
# df.loc[df['Type 1']=='Grass']

# filter using more than  column
# df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison')]

# filter either or on two columns
# df.loc[(df['Type 1']=='Grass') | (df['Type 2']=='Poison')]

# filter on three columns adding in numerical filter
new_df = df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP'] > 70)]

# save filtered list as checkpoint
# new_df.to_csv('filtered.csv',index=False)

# remove old index from filtered, 'drop=true' is to not save the old index as a column at the beginning
# new_df=new_df.reset_index(drop=True)
# to do inplace to reduce memory use
new_df.reset_index(drop=True,inplace=True)

# filter out all the names with mega,"~" is basically "!"
df.loc[~df['Name'].str.contains('Mega')]

# new_df
```
